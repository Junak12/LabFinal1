```python
#BA1A
def Count(text, pat):
    count = 0
    for i in range(len(text) - len(pat) + 1):
        if text[i:i+len(pat)] == pat:
            count += 1
    return count

text = input()
pat = input()
print(Count(text, pat))
```

```python
#BA1B
def freq(text, k):
    amnt = {}
    for i in range(len(text) - k + 1):
        word = text[i:i+k]
        if word in amnt:
            amnt[word] += 1
        else:
            amnt[word] = 1
    mx = max(amnt.values())
    result = []
    for word in amnt:
        if amnt[word] == mx:
            result.append(word)
    return result

text = input()
k = int(input())
print(*freq(text, k))
```

```python
#BA1F
def result(text):
    skew = [0]
    for i in range(len(text)):
        if text[i] == 'G':
            skew.append(skew[-1] + 1)
        elif text[i] == 'C':
            skew.append(skew[-1] - 1)
        else:
            skew.append(skew[-1])
    mn = min(skew)
    pos = [i for i in range(len(skew)) if skew[i] == mn]
    return pos

text = input()
print(*result(text))
```

```python
#BA1H
def calc(p, q):
    count = 0
    for i in range(len(p)):
        if p[i] != q[i]:
            count += 1
    return count

def result(pat, text, d):
    pos = []
    k = len(pat)
    for i in range(len(text) - k + 1):
        wind = text[i:i+k]
        if calc(pat, wind) <= d:
            pos.append(i)
    return pos

pat = input()
text = input()
d = int(input())
print(*result(pat, text, d))
```

```python
#BA1I
def hamming_distance_two_strings(p_text, q_text):
    return sum(1 for a, b in zip(p_text, q_text) if a != b)

def neighbors_of_pattern(pattern_text, d_value):
    if d_value == 0:
        return {pattern_text}
    if len(pattern_text) == 1:
        return {'A', 'C', 'G', 'T'}
    result_set = set()
    suffix_neighbors = neighbors_of_pattern(pattern_text[1:], d_value)
    for suffix in suffix_neighbors:
        if hamming_distance_two_strings(pattern_text[1:], suffix) < d_value:
            for nuc in 'ACGT':
                result_set.add(nuc + suffix)
        else:
            result_set.add(pattern_text[0] + suffix)
    return result_set

def frequent_words_mismatch(text_input, k_value, d_value):
    freq_dict = {}
    for i in range(len(text_input) - k_value + 1):
        pattern = text_input[i:i+k_value]
        for neighbor in neighbors_of_pattern(pattern, d_value):
            freq_dict[neighbor] = freq_dict.get(neighbor, 0) + 1
    max_count = max(freq_dict.values())
    return [p for p in freq_dict if freq_dict[p] == max_count]

text_input = input()
k_value, d_value = map(int, input().split())
print(*frequent_words_mismatch(text_input, k_value, d_value))

```

```python
#BA1J
def hamming_distance_two_strings(p_text, q_text):
    return sum(1 for a, b in zip(p_text, q_text) if a != b)

def neighbors_of_pattern(pattern_text, d_value):
    if d_value == 0:
        return {pattern_text}
    if len(pattern_text) == 1:
        return {'A', 'C', 'G', 'T'}
    result_set = set()
    suffix_neighbors = neighbors_of_pattern(pattern_text[1:], d_value)
    for suffix in suffix_neighbors:
        if hamming_distance_two_strings(pattern_text[1:], suffix) < d_value:
            for nuc in 'ACGT':
                result_set.add(nuc + suffix)
        else:
            result_set.add(pattern_text[0] + suffix)
    return result_set

def reverse_complement_string(pattern_text):
    comp_dict = {'A':'T', 'T':'A', 'C':'G', 'G':'C'}
    return ''.join(comp_dict[n] for n in reversed(pattern_text))

def frequent_words_mismatch_rc(text_input, k_value, d_value):
    freq_dict = {}
    for i in range(len(text_input) - k_value + 1):
        pattern = text_input[i:i+k_value]
        for neighbor in neighbors_of_pattern(pattern, d_value):
            freq_dict[neighbor] = freq_dict.get(neighbor, 0) + 1
            rev_neighbor = reverse_complement_string(neighbor)
            freq_dict[rev_neighbor] = freq_dict.get(rev_neighbor, 0) + 1
    max_count = max(freq_dict.values())
    return [p for p in freq_dict if freq_dict[p] == max_count]

text_input = input()
k_value, d_value = map(int, input().split())
print(*frequent_words_mismatch_rc(text_input, k_value, d_value))
```

```python
#BA1K
def calc(pat):
    val = {'A':0, 'C':1, 'G':2, 'T':3}
    number = 0
    for char in pat:
        number = 4 * number + val[char]
    return number

def result(text, k):
    freq = [0] * (4**k)
    for i in range(len(text) - k + 1):
        pat = text[i:i+k]
        index = calc(pat)
        freq[index] += 1
    return freq

text = input()
k = int(input())
print(*result(text, k))
```
